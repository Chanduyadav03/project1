```python
#positional arguments
#task 1
def greet(name,age):
  print(f"hello{name},you are {age} years old.")
greet("chandu",23)
```

```python
#keyword arguments
#task 2
def greet(name,age,id):
  print(f"hello{name},you are {age} years old.student{id}")
greet(age=23,name="chandu",id=7708)
```

```python
#default argumnts
#Task 3
def greet(name,age=23):
  print(f"hello{name},you are {age} years old.")
greet("chandu")
greet("sneha",22)
```

```python
#variable-length arguments
#Task 4
def print_numbers(*args):
  for number in args:
    print(number)
print_numbers(0,12,34,56,78,90)
```

```python
#keyword variable-length arguments(**kwargs)
#Task 5
def print_info(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")


print_info(name="chandu",profession="AI ENGINEER")
print_info(country= "india",population=1.49,capital="new delhi")
```

```python
#mixing differnt argument types
#Task 6
def mixed_arguments(a, b=4, *args, **kwargs):
    print(f"a: {a}")
    print(f"b: {b}")
    print(f"args:", args)
    print(f"kwargs:", kwargs)

mixed_arguments(1,2,3,4,5, y = 10 , s =30)
```
