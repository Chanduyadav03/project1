```python
#LAB 33
#Task 1: creating a tuple
c_tuple =(10,20,30,40,50,60,70)
print("tuple:",c_tuple)
```

```python
#Task 2:accessing elements in a tuple
print("first :",c_tuple[0])
print("end :",c_tuple[-1])
```

```python
#Task3:slicing a tuple

print("slice [1:4]:",c_tuple[1:4])
print("slice starting in 3:",c_tuple[:3])
print("slice ending in 2:",c_tuple[2:])
```

```python
#task 4 : concatenating two tuples4
c_tuple =(10,20,30,40,50,60,70)
extra_tuple =(80,90,91)
concatenating=c_tuple+extra_tuple
print("concatenated tuple:",concatenating)
```

```python
# Task  5 :iterating through a tuple
for item in c_tuple:
    print("Item:",item)
```

```python
#Task 6 :membrship test a tuple
print("Is 30 in tuple ?",30 in c_tuple)
print("Is 100 not in tuple ?",100 not in c_tuple)
```

```python
#Task 7 :nested tuples and Accessing Their Elements
nested_tuple = (c_tuple,(70,80,91))
print("Nested tuple:", nested_tuple)
print("accessing nested element (c_tuple[2]):",nested_tuple[0][2])
print("Accessing nested element(90):", nested_tuple[1][2])

```

```python
# Task 8 :Applying built-in Functions on tuples
print("Length of tuple :",len(c_tuple))
print("Maximum value :",max(c_tuple))
print("minimum value:",min(c_tuple))
print("sum of elements:",sum(c_tuple))
```

```python
#Task 9:updating a tuple (using list conversion)
temp_list = list(c_tuple)
temp_list[2] =44
updated_tuple = tuple(temp_list)
print("updated tuple :",updated_tuple)
```

```python
#Task 10:deleting a tuple
del c_tuple
print("c_tuple is deleted")
```
