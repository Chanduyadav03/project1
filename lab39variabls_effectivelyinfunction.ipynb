```python
#task 1
#local scope
def local_scop_example():
  x=10
  print("local variable x :",x)
local_scop_example()
```

```python
#task 2
#global scope
x=20
def global_scope_example():
  print("global variable x :",x)
global_scope_example()
print("global variable x outside function:",x)
```

```python
#task 3
#enclosing scope
def outer_function():
  y=30
  def inner_function():
    print("enclosing variable y:",y)
  inner_function()

outer_function()


```

```python
#task4
#modifing global variables
x=10
def modify_global_variable():
  global x
  x=20
modify_global_variable()
print("global x:",x)
```
